# Task 4: One-to-one NMI filter

```python
from pathlib import Path
import json

import pandas as pd

# Set source and output paths.
task_dir = Path("Task")
mapping_path = task_dir / "4" / "building_nmi_mapping.json"
summary_path = task_dir / "EDA_SummaryByNMI.csv"
output_path = task_dir / "EDA_1to1 NMI-building.csv"

# Load mapping and EDA summary.
with mapping_path.open() as f:
    mapping = json.load(f)

summary = pd.read_csv(summary_path)
summary = summary.rename(columns={summary.columns[0]: "nmi"})

# Keep buildings mapped to exactly one NMI.
one_to_one = pd.DataFrame(
    [
        {"nmi": nmis[0], "building": building}
        for building, nmis in mapping["building_to_nmis"].items()
        if len(nmis) == 1
    ]
)

# Join the one-to-one labels to EDA rows.
result = summary.merge(one_to_one, on="nmi", how="inner")
result = result[["nmi", "building", *summary.columns.drop("nmi")]]

# Save the filtered CSV.
result.to_csv(output_path, index=False)
print(f"Saved {len(result)} rows to {output_path}")
```
