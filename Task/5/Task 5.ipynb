```python
# Plotting the average daily pattern for weekdays vs weekends for each column in the aggregated DataFrame
week = agg_df.groupby([agg_df.index.dayofweek < 5, agg_df.index.hour]).mean()

for col in agg_df:
    plt.figure(figsize=(4,3))
    plt.plot(week[col].loc[True], label = 'weekday') # Plot the average consumption for each hour of the day for weekdays
    plt.plot(week[col].loc[False], label = 'weekend') # Plot the average consumption for each hour of the day for weekends

    plt.xlabel('hour')
    plt.ylabel('consumption')
    plt.title(f"{col.replace(' consumption', '')} weekday vs weekend")
    plt.legend()
    plt.xticks(np.arange(0, 24, 2)) # Set x-ticks to be every 2 hours of the day
    
    plt.show()
```
